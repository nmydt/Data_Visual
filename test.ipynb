```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.sans-serif'] = ['SimHei']

labels = ['周一','周二','周三','周四','周五','周六','周日']
def getarr(a):
    a11 = []
    for i in a:
        a11.append(int(i))
    return a11
# 早饭
a1 = getarr(np.random.randint(1, 5, (7,1)))
#午饭
a2 = getarr(np.random.randint(10, 20, (7,1)))
# 晚饭
a3 = getarr(np.random.randint(5, 10, (7,1)))


x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width*2, a1, width, label='早餐')
rects2 = ax.bar(x -width+0.01, a2, width, label='午餐')
rects3 = ax.bar(x +0.02, a3, width, label='晚餐')
# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('消费金额/元')
ax.set_xlabel('一周')
ax.set_title('B19042015代雪刚 一周餐费情况')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
fig.tight_layout()

plt.show()
```

# Pyecharts

```python
# -*- coding: utf-8 -*-

#声明Notebook类型，必须在引入pyecharts.charts等模块前声明
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_NOTEBOOK

from pyecharts import options as opts
from pyecharts.charts import Funnel, Page
import pymysql

#连接Hadoop数据库
v1 = []
v2 = []
conn = pymysql.connect(host='lylg102',port=3306,user='root',password='000000',db='test',charset='utf8')
cursor = conn.cursor()
#读取Hadoop数据
sql_num = "SELECT province,ROUND(SUM(profit),2) FROM orders WHERE dt=2019 and region='华东' GROUP BY province"
cursor.execute(sql_num)
sh = cursor.fetchall()
for s in sh:
    v1.append(s[0])
    v2.append(s[1])
    
#画漏斗图
def funnel_label() -> Funnel:
    c = (
        Funnel()
        .add("利润额",
            [list(z) for z in zip(v1, v2)],
            sort_="descending",               #默认是sort_="descending"即从大到小，也可以设置为ascending即反向漏斗
            label_opts=opts.LabelOpts(position="inside"),
        )
        .set_global_opts(title_opts=opts.TitleOpts(title="区域利润额比较分析", subtitle="2019年企业经营状况"),
                         toolbox_opts=opts.ToolboxOpts(),
                         legend_opts=opts.LegendOpts(is_show=True)
                        )
    )
    return c

#第一次渲染时候调用load_javasrcript文件
funnel_label().load_javascript()
#展示数据可视化图表
funnel_label().render_notebook()
```

```python
# -*- coding: utf-8 -*-

#声明Notebook类型，必须在引入pyecharts.charts等模块前声明
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_NOTEBOOK

from pyecharts import options as opts
from pyecharts.charts import Page, Radar
import pymysql

v2 = ['80','85','90','88','75','78','80','82']
#画雷达图
def radar_base() -> Radar:
    c = (
        Radar()
        .add_schema(
            schema=[
                opts.RadarIndicatorItem(name="数据挖掘技术与应用", max_=100),
                opts.RadarIndicatorItem(name="数据可视化技术", max_=100),
                opts.RadarIndicatorItem(name="大数据分析与内存计算", max_=100),
                opts.RadarIndicatorItem(name="人工智能技术", max_=100),
                opts.RadarIndicatorItem(name="云计算技术", max_=100),
                opts.RadarIndicatorItem(name="形势与政策", max_=100),
                opts.RadarIndicatorItem(name="职业生涯与创新创业思维", max_=100),
                opts.RadarIndicatorItem(name="平均成绩", max_=100),
            ]
        )
        .add("成绩", [v2])
        .set_global_opts(
            title_opts=opts.TitleOpts(title="B19042015代雪刚 ", subtitle="本学期课程成绩"),
            legend_opts=opts.LegendOpts(selected_mode="single"),
            toolbox_opts=opts.ToolboxOpts()
        )
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    )
    return c

#第一次渲染时候调用load_javasrcript文件
radar_base().load_javascript()
#展示数据可视化图表
radar_base().render_notebook()
```

```python
# -*- coding: utf-8 -*-

#声明Notebook类型，必须在引入pyecharts.charts等模块前声明
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_NOTEBOOK

from pyecharts import options as opts
from pyecharts.charts import Sunburst

def sunburst() -> Sunburst:
    data = [
        opts.SunburstItem(
            name="爷爷代一",
            value=75,
            children=[
                
                opts.SunburstItem(
                    name="爸爸代二",
                    value=42,
                    children=[
                        opts.SunburstItem(name="我代雪刚", value=21),
                        opts.SunburstItem(name="哥哥代三", value=23),
                    ],
                ),
            ],
        ),

    ]

    c = (
        Sunburst()
        .add(series_name="", data_pair=data, radius=[0, "90%"])
        .set_global_opts(title_opts=opts.TitleOpts(title="B19042015代雪刚的家庭树旭日图"),
                         toolbox_opts=opts.ToolboxOpts())
        .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}"))
    )
    return c

#第一次渲染时候调用load_javasrcript文件
sunburst().load_javascript()
#展示数据可视化图表
sunburst().render_notebook()
```
